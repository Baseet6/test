```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Load the dataset
file_path = "https://raw.githubusercontent.com/Baseet6/test/refs/heads/main/ai_job_market_insights.csv"
df = pd.read_csv(file_path)

# Display first few rows to understand the structure
df.head()



```

```python
# 2. Box Plot: Salary Distribution for AI vs. Non-AI Jobs
plt.figure(figsize=(18, 5))
sns.boxplot(data=df, x="AI_Adoption_Level", y="Salary_USD", palette="coolwarm")
plt.xlabel("AI Adoption Level", fontsize=12)
plt.ylabel("Salary Distribution (USD)", fontsize=12)
plt.title("Salary Distribution: AI vs. Non-AI Jobs", fontsize=14)
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.show()


```

```python
# Fixing the graph for AI's impact on remote jobs
plt.figure(figsize=(7, 5))
sns.barplot(data=df, x="Remote_Friendly", y="Salary_USD", hue="Remote_Friendly", palette="viridis", estimator=lambda x: sum(x)/len(x))
plt.xlabel("Remote Work Option")
plt.ylabel("Average Salary (USD)")
plt.title("Impact of AI on Remote Jobs Salaries")
plt.show()



```

```python
# 4. Pie Chart: AI Job Adoption Across Industries
plt.figure(figsize=(6, 6))
industry_counts = df["Industry"].value_counts()
plt.pie(industry_counts, labels=industry_counts.index, autopct='%1.1f%%', colors=sns.color_palette("coolwarm", len(industry_counts)))
plt.title("AI Job Adoption Across Industries", fontsize=14)
plt.show()

# 5. Table: Required Skills for AI Jobs
ai_related_jobs = df[df["AI_Adoption_Level"] == "High"]
required_skills_table = ai_related_jobs["Required_Skills"].value_counts().reset_index()
required_skills_table.columns = ["Skill", "Count"]

```
